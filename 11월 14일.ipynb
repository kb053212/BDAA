## 3. 데이터 전처리

##### 1월

대부분 결측치로만 이루어져 있는 진행상태(약식), 예외처리구분, 배송회차 컬럼 삭제

```python
df1.drop(['진행상태 (약식)','예외처리구분','배송회차'], axis = 1, inplace = True)
```

```python
df1.isna().sum()
```

```python
# 아이디 컬럼

df1['아이디'].fillna(value = '미입력', inplace = True)
# 아이디가 누락된 데이터를 '미입력'으로 대체
```

```python
df1['아이디'].isna().sum()
```

```python
# 전시유입코드

df1['전시유입코드'].value_counts()
```

```python
df1['전시유입코드'].fillna(value = 'CHT', inplace = True)
# 전체 데이터에 미치는 영향을 줄이기 위해 상대적으로 빈도수가 높은 CHT로 대체
```

```python
df1.isna().sum()
```

```python
#주문완료일시

x = df1[df1['주문완료일시'].isna() == True]
x[['구매확정','주문완료일시']]
# 주문완료일시가 누락된 고객은 모두 구매미확정 고객임
```

```python
df1['주문완료일시'].fillna(value = '구매 미확정', inplace = True)
# 결측치를 다른 구매미확정 고객들의 평균으로 대체하는 법 찾아보기
```

```python
df1.isna().sum()
```

```python
#구매확정일시

x = df1[df1["구매확정"] == '구매확정']
x_1 = x[['구매확정','구매확정일시']]
x_1.isna().sum()
#구매확정 고객 중에 구매확정일시가 누락된 고객은 없음
#따라서 결측치를 "구매미확정" 으로 대체
```

```python
df1['구매확정일시'].fillna(value = '구매 미확정', inplace = True)
```
