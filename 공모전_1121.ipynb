```python
df_category = df_comb[['주문번호','표준카테고리', '수량', '월', '일', '요일']]
```

```python
# 카테고리를 세분화하기 - 대/중/소 분류
big_list = []
mid_list = []
small_list = []
for i in range(len(df_category['표준카테고리'])):
    big, mid, small = df_category['표준카테고리'][i].split('>')
    big_list.append(big.strip())
    mid_list.append(mid.strip())
    small_list.append(small.strip())
```

```python
df_category['대분류'] = big_list
df_category['중분류'] = mid_list
df_category['소분류'] = small_list
df_category
```

```python
df_category.sort_values('대분류', inplace =True)
```

```python
# 월별 카테고리별 주문건 수 확인
month_big = df_category.pivot_table(index = '월', columns = '대분류', values = '주문번호', aggfunc = 'count')
month_big
```

```python
month_big_sum = df_category.pivot_table(index = '월', columns = '대분류', values = '수량', aggfunc = 'sum')
month_big_sum
```
