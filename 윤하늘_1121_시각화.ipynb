## EDA

### 시간대 vs 주문건수

```python
mt_num = pd.DataFrame(columns = ['월', '시간대', '평균주문건수'])
```

```python
# 각 월마다 시간대별 평균주문건수를 나타내는 데이터프레임 생성
for m in np.arange(1, 10):
    for t in np.arange(0, 24):
        d_ct = df_fin[(df_fin['주문접수_월'] == m) & (df_fin['주문접수_시간대'] == t)].groupby(['주문접수_일']).agg(ct = ('상품주문번호', 'count'))
        num = round(d_ct['ct'].mean(), 2)
        mt_num = mt_num.append(pd.DataFrame([[m, t, num]], columns = ['월', '시간대', '평균주문건수']))
```

```python
mt_num = mt_num.reset_index(drop = True)
mt_num
```

```python
fig = plt.figure(figsize = (15, 10))

sns.lineplot(x = '시간대', y = '평균주문건수', hue = '월', data = mt_num, palette = 'bright')

plt.ylim(0, 130)
plt.xticks(np.arange(0, 24),
          labels = ['0시', '', '2시', '', '4시', '', '6시', '', '8시', '', '10시', '',
                    '12시', '', '14시', '', '16시', '', '18시', '', '20시', '', '22시', ''])

plt.tick_params(axis = 'x', labelsize = 13)
plt.tick_params(axis = 'y', labelsize = 12)

plt.xlabel('시간대', labelpad = 10, fontdict = {'size' : 12}, loc = 'right')
plt.ylabel('평균주문건수', labelpad = 10, fontdict = {'size' : 12}, loc = 'top', rotation = 0)

plt.legend(loc = 'upper left', ncol = 5, title = '월')
plt.title('시간대에 따른 평균주문건수의 변화', pad = 15, size = 20)
```

> 전체적으로 0시부터 4시까지는 감소, 10시에서 11시까지는 증가하다가 다시 감소하는 경향이 있다.          
  그 이후로 비슷한 주문건수를 유지하다가 20시 이후로 다시 증가하는 것을 알 수 있다.

> 다른 달에 비해 5, 6, 7, 8월에는 시간대별로 주문건수의 변화의 폭이 크다.

> 특히 8월에는 21시에 주문량이 급증한 것을 볼 수 있다.

### 유형 vs 주문건수

```python
df_fin.query("예외처리구분 == '구매확정후취소'")['유형']
# 구매확정후취소한 주문에 대해서는 유형이 '주문'으로 구분되어 있음을 확인
```

```python
# 취소되거나 반품된 주문에 대한 금액 0으로 처리
df_fin2 = df_fin.copy()
df_fin2.loc[(df_fin['유형'] == '취소(주문취소)') | (df_fin['유형'] == '반품') | (df_fin['예외처리구분'] == '구매확정후취소'), '주문금액'] = 0
```

```python
# 선물하기와 일반주문의 월별 주문건수, 총액 데이터프레임 생성
df_fin3 = df_fin2.groupby(['주문구분', '주문접수_월']).agg(주문건수 = ('주문금액', 'count'), 총액 = ('주문금액', 'sum'))
df_fin3 = df_fin3.reset_index()
df_fin3
```

```python
fig = plt.figure(figsize = (15, 6))
plt.subplots_adjust(wspace = 0.3)

ax1 = fig.add_subplot(1, 2, 1)
sns.barplot(x = '주문접수_월', y = '주문건수', data = df_fin3.query("주문구분 == '일반주문'"), alpha = 0.5, ax = ax1)
ax1.set_xlabel('월')
ax1.set_ylabel('주문건수', loc = 'top', rotation = 0)

ax2 = ax1.twinx()
sns.lineplot(x = ax1.get_xticks(), y = '총액', data = df_fin3.query("주문구분 == '일반주문'"), ax = ax2)
ax2.set_ylabel('총액', loc = 'top', rotation = 0)
ax2.set(yticks = [0,100000000, 200000000, 300000000, 400000000, 500000000, 600000000, 700000000, 800000000],
        yticklabels = ['0', '1억', '2억', '3억', '4억', '5억', '6억', '7억', '8억'])
ax2.set_title('월별 일반주문 주문건수와 총액')


ax3 = fig.add_subplot(1, 2, 2)
sns.barplot(x = '주문접수_월', y = '주문건수', data = df_fin3.query("주문구분 == '선물하기'"), alpha = 0.5, ax = ax3)
ax3.set_xlabel('월')
ax3.set_ylabel('주문건수', loc = 'top', rotation = 0)

ax4 = ax3.twinx()
sns.lineplot(x = ax3.get_xticks(), y = '총액', data = df_fin3.query("주문구분 == '선물하기'"), ax = ax4)
ax4.set_ylabel('총액', loc = 'top', rotation = 0)
ax4.set(yticks = [0,1000000, 2000000, 3000000, 4000000, 5000000, 6000000, 7000000, 8000000],
        yticklabels = ['0', '1천만원', '2천만원', '3천만원', '4천만원', '5천만원', '6천만원', '7천만원', '8천만원'])
ax4.set_title('월별 선물하기 주문건수와 총액')
```

> 월별 주문건수와 총액은 유사한 변화의 양상을 나타낸다.

> 일반주문과 선물하기 모두 2, 3, 4월에서는 낮은 주문건수와 총액을, 5, 6, 8월에 높은 주문건수와 총액을 보이고 있다.

### 회원구분 vs 주문건수

```python
df_fin6 = df_fin2.groupby(['회원구분', '주문접수_월']).agg(주문건수 = ('주문금액', 'count'))
df_fin6 = df_fin6.reset_index().set_index('회원구분')
df_fin6
```

```python
plt.figure(figsize = (20,20))
color = ['#ABDEE6', '#CBAACB', '#FFCCB6']

for month in np.arange(1, 10):
    ax = plt.subplot(grid[month - 1])
    df_fin6[df_fin6['주문접수_월'] == month].sort_values('주문건수', ascending = False).iloc[:3].plot.pie(x = '회원구분', y = '주문건수', 
                                                 explode = [0, 0.1, 0.1], autopct = '%1.2f%%', colors = color, ax = ax)
    # 주문건수가 많은 상위 3개의 회원만 파이 차트 그림
    ax.set_title(f'{month}월에 주문한 회원 비율')
    ax.set_ylabel('')
plt.show()
```
