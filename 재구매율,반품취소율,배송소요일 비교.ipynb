```python
df['브랜드'] = np.zeros_like(df['주문번호'])
df['브랜드']
```

```python
brand_list = ['비비고', '햇반', '고메', '백설','더건강한', '다시다', 
             '해찬들', '스팸', '삼호어묵', '하선정', '행복한콩', '다담',
             '산들애', '메티에', 'BYO', 'Returnup', '쁘띠첼',
             '맥스봉', '맛밤', '한뿌리', '즐거운동행', '쉐프솔루션', '명가']
no_list = []
for i in range(len(df['전시상품명'])):
    for j in range(len(brand_list)):
        if brand_list[j] in df['전시상품명'][i]:
            df['브랜드'][i] = brand_list[j]
        else:
            pass
    if df['브랜드'][i] == 0:
        no_list.append(i)
        
#전시상품명의 요소 명에 brand_list 요소가 포함되어 있으면 해당 요소의 브랜드를 brand_list로 설정
#브랜드에 추가되지 않았다면 
```

#### 9) L.point 지급 이벤트가 재구매율에 영향을 미치는지

```python
# 전시상품명에 Lpoint 관련 문자가 포함되어 있으면 Lp_list에 추가

Lp_list = []

for i in range(len(df['전시상품명'])) :
    if 'L.point' in df['전시상품명'][i] :
        Lp_list.append(df['전시상품명'][i])
    elif 'Lpoint' in df['전시상품명'][i] :
        Lp_list.append(df['전시상품명'][i])
    elif '엘포인트' in df['전시상품명'][i] :
        Lp_list.append(df['전시상품명'][i])
    else :
        pass
```

```python
Lp_list
```

```python
len(Lp_list)
```

```python
# Lpoint 프로모션 컬럼을 생성
df_cp['Lpoint 프로모션'] = np.zeros_like(df_cp['전시상품명'])

#전시상품명이 Lp_list에 포함되어 있으면 Lpoint프로모션 컬럼에 o, 그렇지 않으면 x
for i in range(len(df['전시상품명'])) :
    if df_cp['전시상품명'][i] in Lp_list :
        df_cp['Lpoint 프로모션'][i] = 'o'
    else :
        df_cp['Lpoint 프로모션'][i] = 'x'
```
