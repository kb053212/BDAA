```python
df_comb = pd.concat([df1, df2, df3, df4, df5,
                     df6.drop(['롯데맴버스오프라인고객번호', '롯데맴버스온라인고객번호'], axis = 1),
                     df7.drop(['롯데맴버스오프라인고객번호', '롯데맴버스온라인고객번호'], axis = 1),
                     df8.drop(['회원번호'], axis = 1), df9.drop(['회원번호'], axis = 1)], ignore_index = True)
# 공통적으로 존재하는 컬럼에 대해 병합
```

```python
df_comb['총액'] = df_comb['판매가'] * df_comb['수량']
```

```python
# 날짜와 관련된 컬럼의 값을 날짜형으로 변환
df_comb['주문접수일시'] = pd.to_datetime(df_comb['주문접수일시'])
df_comb['주문완료일시'] = pd.to_datetime(df_comb['주문완료일시'])
df_comb['구매확정일시'] = pd.to_datetime(df_comb['구매확정일시'])
df_comb['출고지시일'] = pd.to_datetime(df_comb['출고지시일'])
df_comb['상태처리일시'] = pd.to_datetime(df_comb['상태처리일시'])
```

```python
print(df_comb['유입매체'].value_counts(),
      df_comb['주문구분'].value_counts(),
      df_comb['회원구분'].value_counts(),
      df_comb['유형'].value_counts(),
      df_comb['배송회수구분'].value_counts(),
      df_comb['진행단계'].value_counts(),
      df_comb['구매확정'].value_counts(),
      df_comb['예외처리구분'].value_counts(),
      df_comb['배송상품유형'].value_counts(),
      df_comb['배송처리유형'].value_counts(),
      df_comb['배송수단'].value_counts(),
      df_comb['배송예약구분'].value_counts(),
      df_comb['추가옵션'].value_counts(),
      df_comb['주문판매유형'].value_counts(),
      df_comb['상품유형'].value_counts(),
      df_comb['브랜드'].value_counts(), sep = '\n\n')
```

```python
df_fin = df_comb.drop(['진행상태 (약식)', '배송회차', '모델번호', '배송상품유형', '배송처리유형', '배송수단',
                       '배송예약구분', '추가옵션', '주문판매유형', '상품유형', '브랜드'], axis = 1)
# 값이 존재하지 않거나 하나의 값만 존재하는 컬럼 삭제
# 브랜드는 씨제이제일제당, CJ제일제당 같은 것으로 보고 삭제
```
